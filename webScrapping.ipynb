```python
def downloadWebsite(url):
    import bs4 as bs
    import requests
    r = requests.get(url)
    print(r.status_code)
    if r.ok:
        html = bs.BeautifulSoup(r.text)
        
        print(html.prettify())
```

```python
downloadWebsite("http://www.uva.nl")
#https://en.wikipedia.org/wiki/List_of_HTTP_status_codes
```
